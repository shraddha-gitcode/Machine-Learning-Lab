```python
#prime number
    def is_prime(num):
        if num <= 1:
            return False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True

    number = 17
    if is_prime(number):
        print(number, "is a prime number.")
    else:
        print(number, "is not a prime number.")
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("used_cars_data.csv")

# Check columns
print(df.columns)

# If 'Price_log' not present, create it
if 'Price_log' not in df.columns:
    df['Price_log'] = np.log(df['Price'])

fig, axarr = plt.subplots(4, 2, figsize=(12, 18))

sns.stripplot(data=df, x='Location', y='Price_log', ax=axarr[0][0], jitter=True)
axarr[0][0].set_title("Location Vs Price_log")

sns.stripplot(data=df, x='Transmission', y='Price_log', ax=axarr[0][1], jitter=True)
axarr[0][1].set_title("Transmission Vs Price_log")

sns.stripplot(data=df, x='Fuel_Type', y='Price_log', ax=axarr[1][0], jitter=True)
axarr[1][0].set_title("Fuel_Type Vs Price_log")

sns.stripplot(data=df, x='Owner_Type', y='Price_log', ax=axarr[1][1], jitter=True)
axarr[1][1].set_title("Owner_Type Vs Price_log")

sns.stripplot(data=df, x='Brand', y='Price_log', ax=axarr[2][0], jitter=True)
axarr[2][0].set_title("Brand Vs Price_log")

sns.stripplot(data=df, x='Model', y='Price_log', ax=axarr[2][1], jitter=True)
axarr[2][1].set_title("Model Vs Price_log")

sns.stripplot(data=df, x='Seats', y='Price_log', ax=axarr[3][0], jitter=True)
axarr[3][0].set_title("Seats Vs Price_log")

sns.stripplot(data=df, x='Car_Age', y='Price_log', ax=axarr[3][1], jitter=True)
axarr[3][1].set_title("Car_Age Vs Price_log")

plt.tight_layout()
plt.show()
```
