```python
#calculator 

a = float(input("Enter your first number: "))
b = float(input("Enter your second number: "))

print("\nSelect the operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Choose the operation you want to perform: "))

match choice:
    case 1:
        print("Result:", a + b)
    case 2:
        print("Result:", a - b)
    case 3:
        print("Result:", a * b)
    case 4:
        if b != 0:
            print("Result:", a / b)
        else:
            print("Error: Cannot divide by zero.")
    case _:
        print("Invalid choice. Please select between 1 to 4.")



          
```

```python
#prime number
    def is_prime(num):
        if num <= 1:
            return False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True

    number = 17
    if is_prime(number):
        print(number, "is a prime number.")
    else:
        print(number, "is not a prime number.")
```

```python
#bar chart
import numpy as np
import matplotlib.pyplot as plt

a = np.array([20, 140, 45, 60])
b = np.array([10, 20, 50, 70])

plt.bar(a, b, color='teal', width=10) 
plt.xlabel("a values")
plt.ylabel("b values")
plt.title("Bar Chart: a vs b")

plt.show()
```
